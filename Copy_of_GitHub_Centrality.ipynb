```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import os
```

```python
# Path file CSV yang diunggah atau di-drive
csv_path = "/content/musae_git_edges.csv"

# Membaca edge list
edges_df = pd.read_csv(csv_path)
```

**Degree Centrality**

```python
# Membuat graph dari edge list
G = nx.from_pandas_edgelist(edges_df, source="id_1", target="id_2")
```

```python
# Menghitung Degree Centrality
degree_centrality = nx.degree_centrality(G)

# Menampilkan beberapa hasil pertama
print("Sample Degree Centrality Values:")
print(list(degree_centrality.items())[:10])
```

```python
# Mengurutkan node berdasarkan degree centrality dari yang terbesar ke terkecil
sorted_degree_centrality = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)

# Menampilkan hasil
print("Nodes sorted by Degree Centrality:")
for node, centrality in sorted_degree_centrality:
    print(f"Node: {node}, Degree Centrality: {centrality:.4f}")
```

```python
# Menghitung Closeness Centrality
sampled_nodes = list(G.nodes())  # Ambil 100 node pertama
closeness_centrality = {n: nx.closeness_centrality(G, u=n) for n in sampled_nodes}
print(list(closeness_centrality.items())[:10])
```

```python
# Mengurutkan node berdasarkan closeness centrality dari yang terbesar ke terkecil
sorted_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)

# Menampilkan hasil
print("Nodes sorted by Closeness Centrality:")
for node, centrality in sorted_closeness:
    print(f"Node: {node}, Closeness Centrality: {centrality:.6f}")
```

```python
# Menghitung Betweenness Centrality
approx_betweenness = nx.betweenness_centrality(G, k=100)  # Hanya gunakan 100 node sebagai sampel
print(list(approx_betweenness.items())[:10])
```

```python
# Visualisasi Graph (lebih cepat)
plt.figure(figsize=(10, 10))
subset_nodes = list(G.nodes())[:100]  # Mengambil subgraph dari 100 node pertama
subG = G.subgraph(subset_nodes)

pos = nx.kamada_kawai_layout(subG)  # Layout yang lebih cepat dibandingkan spring_layout

# Gambar node dan edges secara terpisah
nx.draw_networkx_nodes(subG, pos, node_size=5, alpha=0.6)
nx.draw_networkx_edges(subG, pos, alpha=0.2)

plt.title("Subgraph Visualization - Git Web ML Dataset (100 Nodes)")
plt.show()

```

```python
# Fungsi untuk menampilkan Top 10 Node berdasarkan centrality tertentu
def top_centrality(centrality_dict, centrality_type):
    sorted_centrality = sorted(centrality_dict.items(), key=lambda x: x[1], reverse=True)
    print(f"\nTop 10 Node dengan {centrality_type} Centrality Tertinggi:")
    for node, centrality in sorted_centrality[:10]:
        print(f"Node {node}: {centrality}")

# Top 10 Degree Centrality
top_centrality(degree_centrality, "Degree")

# Top 10 Closeness Centrality
top_centrality(closeness_centrality, "Closeness")

# Top 10 Betweenness Centrality
top_centrality(approx_betweenness, "Betweenness")
```

```python
# Menghitung jumlah followers (degree) setiap node
followers_count = dict(G.degree())

# Menampilkan 10 node pertama dengan jumlah followers-nya
print("Top 10 jumlah followers:")
top_followers = sorted(followers_count.items(), key=lambda x: x[1], reverse=True)[:10]
for node, followers in top_followers:
    print(f"Node {node}: {followers} followers")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung jumlah followers (degree) setiap node
followers_count = dict(G.degree())

# Mengambil 10 node dengan followers terbanyak dan mengurutkannya
top_followers = sorted(followers_count.items(), key=lambda x: x[1], reverse=True)[:10]
nodes, followers = zip(*top_followers)  # Memisahkan node dan jumlah followers

# Membuat diagram batang dengan urutan dari yang paling banyak ke paling sedikit
plt.figure(figsize=(8, 5))
sns.barplot(x=list(nodes), y=list(followers), color="cornflowerblue", order=list(nodes))

# Menambahkan label
plt.xlabel("Node")
plt.ylabel("Jumlah Followers")
plt.title("Top 10 Node dengan Jumlah Followers Tertinggi")
plt.xticks(rotation=45)  # Memiringkan label agar lebih mudah dibaca

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Fungsi untuk menampilkan Diagram Batang (sorted dari yang tertinggi ke rendah)
def plot_top_centrality(centrality_dict, centrality_type):
    sorted_centrality = sorted(centrality_dict.items(), key=lambda x: x[1], reverse=True)[:10]
    nodes, values = zip(*sorted_centrality)  # Memisahkan node dan nilai centrality

    plt.figure(figsize=(8, 5))
    sns.barplot(x=list(nodes), y=list(values), color="cornflowerblue", order=list(nodes))
    plt.xlabel("Node")
    plt.ylabel(f"{centrality_type} Centrality")
    plt.title(f"10 Pengguna Berpengaruh dengan {centrality_type} Centrality")
    plt.xticks(rotation=45)  # Rotasi label agar lebih terbaca
    plt.show()

# Menampilkan Diagram Batang untuk Degree Centrality
plot_top_centrality(degree_centrality, "Degree")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Pilih node pusat yang ingin divisualisasikan
target_node = 31890

# Tetapkan jumlah tetangga yang ingin ditampilkan di sekitar node pusat
neighbors = list(G.neighbors(target_node))[:300]  # Ambil maksimal 500 tetangga agar visualisasi tetap jelas

# Buat subgraph dengan node pusat dan tetangganya
subgraph = G.subgraph([target_node] + neighbors)

# Plot visualisasi subgraf
plt.figure(figsize=(8, 8))
pos = nx.spring_layout(subgraph, seed=42)  # Layout untuk penyusunan node
nx.draw(subgraph, pos, with_labels=True, node_size=100, node_color="skyblue", edge_color="gray", font_size=10)
nx.draw_networkx_nodes(subgraph, pos, nodelist=[target_node], node_size=1000, node_color="red")  # Warna merah untuk node utama
plt.title(f"Visualisasi Subgraf dari Node {target_node}")
plt.show()
```

```python
# Menampilkan jumlah node dan edge
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

print(f"Jumlah Node: {num_nodes}")
print(f"Jumlah Edge: {num_edges}")
```

```python
# Menghitung Density
density = nx.density(G)

# Menghitung Transitivity
transitivity = nx.transitivity(G)

# Menampilkan hasil
print(f"Density: {density}")
print(f"Transitivity: {transitivity}")
```
